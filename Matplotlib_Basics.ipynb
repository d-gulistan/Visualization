<h3>Basic Matplotlib Diagrams</h3>

This notebook provides an overview of plotting basic diagrams in Matplotlib.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
print(matplotlib.__version__)
```

<strong>Starting with the very basics, let's draw a line in a simple diagram from position (0,0) to position (5,50).</strong>

```python
xpoints = np.array([0, 5])
ypoints = np.array([0, 50])

plt.plot(xpoints, ypoints)
plt.show()
```

<strong>We can draw another line from position (1,1) to position (10,20).</strong>

```python
xpoints = np.array([1, 10])
ypoints = np.array([1, 20])

plt.plot(xpoints, ypoints)
plt.show()
```

<strong>If we only want to plot the markers, without a line, we can utilize the shortcut string notation parameter 'o', which means 'rings'.</strong>

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 12])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

<strong>We can plot multiple points by ensuring that we have an equal number of points on both axes.</strong>

```python
xpoints = np.array([1, 2, 3, 6, 8])
ypoints = np.array([3, 8, 2, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

<strong>We can plot the line without specifying the points on the x-axis. In this case, they will get the default values (0,1,2,3,4,..).</strong>

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

<strong>The "marker" argument emphasizes each point with a specific marker, ex. 'o' draws a circle at the points.</strong>

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, marker='o')
plt.show()
```

<strong>This is how we can mark each point with a star.</strong>

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, marker='*')
plt.show()
```

<strong>In order to further specify the marker, we can use the "marker|line|color" sytax.</string>

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, 'o:m')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, 'o-.g')
plt.show()
```

<strong>We can adjust the size of the marker with the argument 'markersize' or 'ms'.</strong>

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, 'o:b', ms=15)
plt.show()
```

<strong>We can set the color of the marker edges with the argument 'markeredgecolor' or 'mec'.</strong>

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, marker='o', ms=15, mec='r')
plt.show()
```

<strong>We can set the color inside the marker edges with the argument 'markerfacecolor' or 'mfc'.</strong>

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, marker='o', ms=15, mfc='y')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, marker='o', ms=15, mec='r', mfc='k')
plt.show()
```

<strong>We may use the Hexadecimal color values or the color names for our color choices.</strong>

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, marker='o', ms=15, mec='#ff0093', mfc='#ff0093')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, marker='o', ms=15, mec='chocolate', mfc='chocolate')
plt.show()
```

<strong>If we want to the change the style of the plotted line, we can use the argument 'linestyle' or 'ls'.</strong>

```python
ypoints = np.array([3, 8, 2, 9])

plt.plot(ypoints, linestyle='dotted')
plt.show()
```

```python
ypoints = np.array([3, 8, 2, 9])

plt.plot(ypoints, 'y', ls='--')
plt.show()
```

<strong>We can utilize the argument 'color' or 'c' to adjust the color of the line.</strong>

```python
ypoints = np.array([3, 8, 2, 9])

plt.plot(ypoints, color='m', ls=':')
plt.show()
```

```python
ypoints = np.array([3, 8, 2, 9])

plt.plot(ypoints, color='lightgreen')
plt.show()
```

<strong>The width of the line can be changed with the argument 'linewidth' or 'lw'.</strong>

```python
ypoints = np.array([3, 8, 2, 9])

plt.plot(ypoints, linewidth='12.5')
plt.show()
```

<strong>It's possible to plot multiple lines by specifying a 'plt.plot()' function for each line, or by adding the points for the x- and y-axis for each line in the same 'plt.plot()' function.</strong>

```python
y1 = np.array([3, 7, 1, 8])
y2 = np.array([6, 2, 7, 3])

plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 7, 1, 8])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 3])

plt.plot(x1, y1, x2, y2)
plt.show()
```

<strong>It's also possible to draw multiple plots in one figure by using the 'subplots()' function.</strong>

```python
#Plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

#The figure will have 1 row, 2 columns, and this plot will be the first one.
plt.subplot(1, 2, 1)
plt.plot(x,y)

#Plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y, color='m')

plt.show()
```

```python
#Plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y, color='chocolate')

#Plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y, color='y')

plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'm')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'y--')

plt.show()
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='Linear')
plt.plot(x, x**2, label='Quadratic')
plt.plot(x, x**3, label='Cubic')

plt.xlabel('Label on x')
plt.ylabel('Label on y')
plt.title("Three lines")

plt.show()
```

<strong>A title can be added to the plot with the 'title()' function.</strong>

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Expenses")

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y, color='m')
plt.title("Revenue")

plt.show()
```

<strong>It's also possible to add a super title to the entire diagram with the 'suptitle()' function.</strong>

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Expenses")

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y, color='orange')
plt.title("Revenue")

plt.suptitle("BALANCE SHEET")

plt.show()
```
